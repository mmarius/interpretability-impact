```python
import os, sys
sys.path.append(os.path.dirname(os.getcwd()))
from utils import *
import pandas as pd
```

```python
df = pd.read_excel('../data/emnlp2022/Accepted-Papers.xls')
df
```

```python
set(df['Acceptance Status'])
```

```python
df = df[df['Acceptance Status'] != 'Accept-Findings'].copy()
df
```

```python
set(df['Track'])
```

```python
new_df = pd.DataFrame()
new_df['id'] = df['Submission ID']
new_df['title'] = df['Title']
new_df['area'] = df['Track']
new_df['source'] = 'EMNLP'
new_df['year'] = 2022
new_df['doi'] = None
new_df['abstract'] = None

new_df
```

```python
# this paper was actually published in EMNLP 2023
new_df = new_df[~(new_df['title'] == 'Federated Meta-Learning for Emotion and Sentiment Aware Multi-modal Complaint Identification')]
new_df
```

# Saving the results

```python
previous_df = pd.read_csv('../data/cl_papers.csv', index_col=0)
previous_df
```

```python
# we'll delete previous rows from the conference
previous_df = previous_df[~((previous_df['source'] == 'EMNLP') & (previous_df['year'] == 2022))]
previous_df
```

```python
updated_df = pd.concat([previous_df, new_df], ignore_index=True)
updated_df
```

```python
updated_df.to_csv('../data/cl_papers.csv')
```
