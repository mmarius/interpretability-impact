```python
import re

DATA_DIR = '../data/acl2019/data/papers/final'

# there are also workshop papers in /data/acl2019/data, but for now i'll
# ignore them
# BlackboxNLP although it was on ACL2019 was not in the folder

data = []
for paper_id in os.listdir(DATA_DIR):
    path = os.path.join(DATA_DIR, paper_id, f"{paper_id}_metadata.txt")
    with open(path) as f:
        raw_title = f.readlines()[1]
        title = re.search(r'#=%=#(.+)$', raw_title).group(1).strip()
    
    data.append({'id': paper_id, 'title': title})

df = pd.DataFrame(data)
df
```
