```python
new_df = pd.DataFrame()
new_df['id'] = df['Submission ID']
new_df['title'] = df['Submission Title']
new_df['area'] = df['Track']
new_df['source'] = 'EMNLP'
new_df['year'] = 2018
new_df['doi'] = None
new_df['abstract'] = None

new_df
```

```python
new_df = new_df.drop_duplicates(subset='id', keep='first')
new_df
```

# Saving the results

```python
previous_df = pd.read_csv('../data/cl_papers.csv', index_col=0)
previous_df
```

```python
# we'll delete previous rows from the conference
previous_df = previous_df[~((previous_df['source'] == 'EMNLP') & (previous_df['year'] == 2018))]
previous_df
```

```python
updated_df = pd.concat([previous_df, new_df], ignore_index=True)
updated_df
```

```python
updated_df.to_csv('../data/cl_papers.csv')
```
