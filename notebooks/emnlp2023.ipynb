```python
import os, sys
sys.path.append(os.path.dirname(os.getcwd()))
from utils import *
import pandas as pd
```

```python
ACCEPTED_PAPERS_SHEET = 6
df = pd.read_excel('../data/emnlp2023/EMNLP 2023 papers and schedule.xlsx', sheet_name=ACCEPTED_PAPERS_SHEET)
df
```

```python
set(df['Decision'])
```

```python
new_df = pd.DataFrame()
new_df['id'] = df['Paper Number']
new_df['title'] = df['Title']
new_df['area'] = df['Submission Track']
new_df['source'] = 'EMNLP'
new_df['year'] = 2023
new_df['doi'] = None
new_df['abstract'] = None

new_df
```

# Saving the results

```python
previous_df = pd.read_csv('../data/cl_papers.csv', index_col=0)
previous_df
```

```python
# we'll delete previous rows from the conference
previous_df = previous_df[~((previous_df['source'] == 'EMNLP') & (previous_df['year'] == 2023))]
previous_df
```

```python
updated_df = pd.concat([previous_df, new_df], ignore_index=True)
updated_df
```

```python
updated_df.to_csv('../data/cl_papers.csv')
```
