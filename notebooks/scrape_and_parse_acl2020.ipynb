```python
import requests
from bs4 import BeautifulSoup

url = 'https://virtual.acl2020.org/papers.json'

response = requests.get(url)

response = response.json()
response
```

```python
data = []

for paper in response:
    title = paper['content']['title']
    track = paper['content']['track']
    paper_type = paper['content']['paper_type']
    paper_id = paper['id']
    
    source = paper_id.split('.')[0]
    if source not in ['tacl', 'main', 'cl']:
        continue
    
    data.append({ 'title': title, 'area': track, 'id': paper_id })

df = pd.DataFrame(data)
df
```

```python
research_areas = {
    'Cognitive Modeling and Psycholinguistics': False,
    'Computational Social Science and Social Media': False,
    'Dialogue and Interactive Systems': False,
    'Discourse and Pragmatics': False,
    'Ethics and NLP': False,
    'Generation': False,
    'Information Extraction': False,
    'Information Retrieval and Text Mining': False,
    'Interpretability and Analysis of Models for NLP': True,
    'Language Grounding to Vision, Robotics and Beyond': False,
    'Machine Learning for NLP': False,
    'Machine Translation': False,
    'NLP Applications': False,
    'Phonology, Morphology and Word Segmentation': False,
    'Question Answering': False,
    'Resources and Evaluation': False,
    'Semantics: Lexical': False,
    'Semantics: Sentence Level': False,
    'Semantics: Textual Inference and Other Areas of Semantics': False,
    'Sentiment Analysis, Stylistic Analysis, and Argument Mining': False,
    'Speech and Multimodality': False,
    'Summarization': False,
    'Syntax: Tagging, Chunking and Parsing': False,
    'Theme': False,
    'Theory and Formalism in NLP (Linguistic and Mathematical)': False
}


df['interpretability'] = df['area'].apply(lambda x: research_areas[x])
df
```
