```python
import os, sys
sys.path.append(os.path.dirname(os.getcwd()))
from utils import *
import pandas as pd
```

```python
df = pd.read_excel('../data/acl2022/presentation slots.xlsx', sheet_name=0,  skiprows=5)
df
```

```python
set(df['Event session'])
```

```python
valid_sessions_pattern = r'^\S+:\s*.+$'
filtered_df = df[df['Event session'].str.match(valid_sessions_pattern, na=False)]
filtered_df = filtered_df.drop_duplicates(subset='Title')
filtered_df = filtered_df.drop_duplicates(subset='Paper number')


filtered_df
```

```python
split_columns = filtered_df['Title'].str.split(':', n=1, expand=True)
filtered_df['source'] = split_columns[0].str.strip()
filtered_df['clean_title'] = split_columns[1].str.strip()

# this drops extra columns they have
filtered_df = filtered_df.drop(columns=[f"Unnamed: {i}" for i in range(6, 17)])

filtered_df
```

```python
set(filtered_df['source'])
```

```python
allowed_sources = ['Short', 'Long', 'CL', 'TACL']
filtered_df = filtered_df[filtered_df['source'].isin(allowed_sources)].copy()

filtered_df
```

```python
split_columns = filtered_df['Event session'].str.split(':', n=1, expand=True)
filtered_df.loc[:, 'area'] = split_columns[1].str.strip()

filtered_df
```

```python
set(filtered_df['area'])
```

```python
research_areas = {
    'Computational Social Science and Cultural Analytics': False,
    'Dialogue and Interactive Systems': False,
    'Discourse and Pragmatics & Ethics in NLP': False,
    'Ethics in NLP': True,
    'Generation': False,
    'Information Extraction': False,
    'Information Retrieval and Text Mining': False,
    'Interpretability and Analysis of Models for NLP': True,
    'Language Groundings, Speech and Multimodality': False,
    'Linguistic Theories, Cognitive Modeling and Psycholinguistics': False,
    'Machine Learning for NLP': False,
    'Machine Translation and Multilinguality': False,
    'NLP Applications': False,
    'Phonology, Morphology and Word Segmentation': False,
    'Question Answering': False,
    'Resources and Evaluation': False,
    'Semantics': False,
    'Sentiment Analysis, Stylistic Analysis, and Argument Mining': False,
    'Special Theme on Language Diversity: From Low Resource to Endangered': False,
    'Summarization': False,
    'Syntax: Tagging, Chunking and Parsing': False
}

filtered_df['interpretability'] = filtered_df['area'].apply(lambda x: research_areas[x])
filtered_df
```

```python
new_df = pd.DataFrame()
new_df['id'] = filtered_df['Paper number']
new_df['title'] = filtered_df['clean_title']
new_df['area'] = filtered_df['area']
new_df['interpretability'] = filtered_df['interpretability']
new_df['source'] = 'ACL2022'

new_df
```

```python
new_df['doi'] = new_df['title'].apply(get_acl_anthology_doi)
new_df
```

# Saving the results

```python
previous_df = pd.read_csv('../data/parsed_data.csv')
previous_df = previous_df.drop('Unnamed: 0', axis=1)
previous_df
```

```python
# we'll delete previous rows from the conference
previous_df = previous_df[previous_df['source'] != 'ACL2022']
previous_df
```

```python
updated_df = pd.concat([previous_df, new_df], ignore_index=True)
updated_df
```

```python
updated_df.to_csv('../data/parsed_data.csv')
```
