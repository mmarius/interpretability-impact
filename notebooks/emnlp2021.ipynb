```python
new_df = pd.DataFrame()
new_df['id'] = None
new_df['title'] = df['Title']
new_df['area'] = df['Track']
new_df['source'] = 'EMNLP'
new_df['year'] = 2021
new_df['doi'] = None
new_df['abstract'] = None

new_df
```

# Saving the results

```python
previous_df = pd.read_csv('../data/cl_papers.csv', index_col=0)
previous_df
```

```python
# we'll delete previous rows from the conference
previous_df = previous_df[~((previous_df['source'] == 'EMNLP') & (previous_df['year'] == 2021))]
previous_df
```

```python
updated_df = pd.concat([previous_df, new_df], ignore_index=True)
updated_df
```

```python
updated_df.to_csv('../data/cl_papers.csv')
```
