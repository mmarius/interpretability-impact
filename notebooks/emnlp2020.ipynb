```python
import pandas as pd

url = 'https://raw.githubusercontent.com/acl-org/emnlp-2020-virtual-conference/master/sitedata/main_papers.csv'
df = pd.read_csv(url)
print(df.head())
```

```python
df['doi'] = '10.18653/v1/2020.emnlp-' + df['UID']
df['area'] = df['track']
df['id'] = df['UID']
df['source'] = 'EMNLP'
df['year'] = 2020
columns_to_keep = ['id', 'title', 'area', 'source', 'year', 'doi', 'abstract']
clean_df = df[columns_to_keep]
clean_df
```

```python
clean_df['area'].value_counts()
```

# Saving the results

```python
previous_df = pd.read_csv('../data/cl_papers.csv', index_col=0)
previous_df
```

```python
# we'll delete previous rows from the conference
previous_df = previous_df[~((previous_df['source'] == 'EMNLP') & (previous_df['year'] == 2020))]
previous_df
```

```python
updated_df = pd.concat([previous_df, clean_df], ignore_index=True)
updated_df
```

```python
updated_df.to_csv('../data/cl_papers.csv')
```
